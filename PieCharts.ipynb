```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C://Users//ASMA//Desktop//covid_statewie_eda//Latest Covid-19 India Status.csv")
```

```python
colors = ['#FFFF00','#ff007f','#e74c3c']
labels = ['Active','Discharged','Deaths']
```

```python
for i in range(36):
    plt.figure(figsize=(8,8))
    plt.title(data.loc[i]['State/UTs'])
    plt.pie([data.loc[i]['Active'],data.loc[i]['Discharged'],data.loc[i]['Deaths']],colors=colors,labels=labels,explode=[0,0.1,0.1],autopct='%0.1f')
    plt.show()
```

```python
plt.figure(figsize=(8,8))
sort_data=data.sort_values(by="Active", ascending=False, axis=0)
labels = data.index
plt.pie(x='Active',data=sort_data[:5],labels='State/UTs',startangle=90,autopct='%.1f%%',)
plt.title("Active Cases in Top 5 states in India", fontsize = 24) 
plt.tight_layout() 
plt.show()
```

```python
plt.figure(figsize=(8,8))
sort_data=data.sort_values(by="Total Cases", ascending=False, axis=0)
labels = data.index
plt.pie(x='Total Cases',data=sort_data[:5],labels='State/UTs',startangle=90,autopct='%.1f%%')
plt.title("Total Cases in Top 5 states in India", fontsize = 24) 
plt.tight_layout() 
plt.show()
```
