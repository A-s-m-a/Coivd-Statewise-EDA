```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("C://Users//ASMA//Desktop//covid_statewie_eda//Latest Covid-19 India Status.csv")
```

```python
plt.figure(figsize=(10,10))
plt.title('Total Cases')
sns.barplot(data=data,x='Total Cases',y='State/UTs',color='red')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Active Cases')
sns.barplot(data=data,x='Active',y='State/UTs',color='blue')
plt.show()
```

```python
plt.figure(figsize=(10,10))
plt.title('Discharged')
sns.barplot(data=data,x='Discharged',y='State/UTs',color='pink')
plt.show()
```

```python
plt.style.use('seaborn')
plt.figure(figsize=(10,10))
plt.title('Deaths')
sns.barplot(data=data,x='Deaths',y='State/UTs',color='green')
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Total Cases',data=data, color="red")
plt.title("Total Cases")
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Active',data=data,color='blue')
plt.title("Active cases in India")
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Discharged',data=data,color='green')
plt.title("Discharged patients in India")
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Deaths',data=data,color='yellow')
plt.title("Death summary in India")
```

```python
plt.figure(figsize=(12,5))
plt.xticks(rotation=90)
sns.pointplot(x='State/UTs',y='Active Ratio',data=data,color='orange')
plt.title("Active Ratio summary in India")
```
