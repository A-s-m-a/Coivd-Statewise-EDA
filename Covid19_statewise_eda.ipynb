```python
plt.figure(figsize=(10,10))
sns.relplot(x = 'Total Cases', y = 'Discharged', hue = 'State/UTs', data = data, )
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize = (10,10))
sns.catplot(data=data, kind="bar", x="State/UTs", y="Total Cases", height=10.27, aspect=40.7/10.27, )
plt.xticks(rotation = 90)
```

```python
plt.figure(figsize=(10,10))
sns.barplot(x='State/UTs',y='Total Cases',palette='CMRmap',data=data)
plt.title("The total cases as per State/UTs are",fontsize=15)
plt.xticks(rotation=90)
```

```python
sns.barplot(x = "State/UTs", y = "Active", data = data[:10], palette='CMRmap', )
plt.xticks(rotation=90)
```

```python
data_active=data.sort_values(by="Active", ascending=False, axis=0)
plt.figure(figsize=(12,7))

AC = data_active.iloc[0:10]
sns.barplot(x = 'Total Cases', y = 'Active', data=AC, hue= 'State/UTs')
plt.title("Top 10 active cases in India")
```

```python
data_deaths=data.sort_values(by="Deaths", ascending=False, axis=0)
plt.figure(figsize=(12,8))

DC = data_deaths.iloc[0:10]
sns.barplot(x = 'Total Cases', y = 'Deaths', data=DC, hue= 'State/UTs')
plt.title("Top 10 states death cases in India")
```

```python
data_deaths=data.sort_values(by="Discharged", ascending=False, axis=0)
plt.figure(figsize=(12,7))

DischargedC = data_deaths.iloc[0:10]
sns.barplot(x = 'Total Cases', y = 'Discharged', data=DischargedC, hue= 'State/UTs')
plt.title("Top 10 states discharged cases in India")
```

```python
fig, ax=plt.subplots(figsize=(20,8))
data.plot(kind = "bar", x = "State/UTs", ax = ax, cmap = "tab20")
```
